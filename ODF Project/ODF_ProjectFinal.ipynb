```python
from shapely.geometry import Point
import folium
import ipywidgets as widgets
from IPython.display import display, clear_output

fire_census_df = pd.read_csv('fire_census_data.csv')

# Create GeoDataFrame
fire_census_df['geometry'] = fire_census_df.apply(lambda row: Point(float(row.Long_DD), float(row.Lat_DD)), axis=1)
geo_data = gpd.GeoDataFrame(fire_census_df, geometry='geometry')
geo_data.set_crs(epsg=4326, inplace=True)

# Create dropdowns for Year and County
year_dropdown = widgets.Dropdown(
    options=sorted(geo_data['FireYear'].unique()),
    description='Year:',
    disabled=False,
)

county_dropdown = widgets.Dropdown(
    options=['All'] + sorted(geo_data['CountyName'].unique().tolist()),
    description='County:',
    disabled=False,
)

```

```python
def display_map(year, county):
    clear_output(wait=True)
    display(year_dropdown, county_dropdown)
    # Filter data
    filtered_data = geo_data[(geo_data['FireYear'] == year)]
    if county != 'All':
        filtered_data = filtered_data[filtered_data['CountyName'] == county]

    # Create map
    folium_map = folium.Map(location=[44.0, -120.5], zoom_start=7)
    for idx, row in filtered_data.iterrows():
        if not row['geometry'].is_empty:
            folium.Marker(
                [row['geometry'].y, row['geometry'].x],
                popup=(
                    f"Fire Name: {row['FireName']}<br>"
                    f"Year: {row['FireYear']}<br>"
                    f"Size: {row['EstTotalAcres']} acres"
                ),
                icon=folium.Icon(color='red', icon='fire')
            ).add_to(folium_map)

    # Display the map in the notebook
    display(folium_map)
```

```python
def on_dropdown_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        display_map(year_dropdown.value, county_dropdown.value) 
```

```python
year_dropdown.observe(on_dropdown_change, names='value')
county_dropdown.observe(on_dropdown_change, names='value')
```

```python
display(year_dropdown, county_dropdown)
display_map(year_dropdown.value, county_dropdown.value)
```

```python
average_fire_size_by_county = fire_census_copy.groupby('CountyName')['EstTotalAcres'].mean().round(3)

print(average_fire_size_by_county)
```

```python
human_lightning_by_county = fire_census_copy.groupby('CountyName')['HumanOrLightning'].apply(lambda x: x.mode()[0])
human_lightning_by_county
```

```python
cause_by_county = fire_census_copy.groupby('CountyName')['CauseBy'].apply(lambda x: x.mode()[0])
cause_by_county
```

```python
general_cause_by_county = fire_census_copy.groupby('CountyName')['GeneralCause'].apply(lambda x: x.mode()[0])
general_cause_by_county
```

```python
# Convert to datetime if necessary
fire_census_copy['Ign_DateTime'] = pd.to_datetime(fire_census_copy['Ign_DateTime'])
fire_census_copy['Control_DateTime'] = pd.to_datetime(fire_census_copy['Control_DateTime'])

# Calculate duration of each fire
fire_census_copy['FireDuration'] = fire_census_copy['Control_DateTime'] - fire_census_copy['Ign_DateTime']

# Calculate average fire duration by county
average_fire_duration_by_county = fire_census_copy.groupby('CountyName')['FireDuration'].mean()

print(average_fire_duration_by_county)
```

# Correlation Matricies using Pearson's R, Spearman, and Kendall Tau's Correlations

```python
corrMatrix = merged_networks.corr()
corrMatrix.head()
```

```python
spearman_corr = merged_networks.corr(method='spearman')

spearman_corr.head()
```

```python
# Calculating Kendall's Tau Correlation
kendall_corr = merged_networks.corr(method='kendall')

kendall_corr.head()
```

### Due to the close correlation measures (by virtue of averaging our metrics for this analysis) it was challenging to be able to differentiate these counties with each other. Kendall's Tau offered a greater variation suggesting it fits our model better. For the remainder of this analysis, we will use Kendall's Tau as our primary correlation matrix to model a heatmap and network map off of. 

```python
corr = kendall_corr
```

```python
# Generate a heatmap of the correlation matrix
# Increase the size of the heatmap.

max_corr = corr[corr != 1.0].max().max()
min_corr = corr[corr != 1.0].min().min()

plt.figure(figsize=(18, 16))  # Adjust the size of the figure

# Adjusting the mask to keep the diagonal
mask = np.triu(np.ones_like(corr, dtype=bool), k=1) 

# Choosing a more contrasting colormap
cmap = sns.diverging_palette(240, 10, n=100, center="light", as_cmap=True)

# Drawing the heatmap with adjusted parameters
ax = sns.heatmap(corr, mask=mask, cmap=cmap, vmin=min_corr, vmax=max_corr, center=0,
                 square=True, annot=True, fmt=".2f", 
                 annot_kws={"size": 9}, cbar_kws={"shrink": 0.75})

# Rotating the tick labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation = 0, fontsize=10)

# Set the title with a larger font and adjust the position
plt.title('Kendall Correlation Matrix', fontsize=20)
plt.xlabel('Oregon County', fontsize=15)
plt.ylabel('Oregon County', fontsize=15)

plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```

### The above map attempts to best differentiate correlations without misrepresenting the true correlation measures for each county relationship. In the future, I would explore additional ways to scale this data to better visualize a heat map like this.

```python
# Convert the correlation matrix DataFrame to a numpy matrix
matrix = corr.values

names = list(corr.index)

# Create a dictionary that maps integer node IDs to their names
mapping = dict(zip(range(len(names)), names))


G = nx.from_numpy_array(matrix)
G = nx.relabel_nodes(G, mapping)  
G.remove_edges_from(nx.selfloop_edges(G))  

# Define three different thresholds
thresholds = [0.50, 0.85, 0.99]

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Loop over each threshold and create a subplot for each network
for i, threshold in enumerate(thresholds):
    G_thresholded = nx.Graph()

    # Add nodes and edges based on threshold
    G_thresholded.add_nodes_from(G)
    for u, v, d in G.edges(data=True):
        if abs(d['weight']) >= threshold:
            G_thresholded.add_edge(u, v, weight=d['weight'])

    # Position nodes using a layout algorithm
    pos = nx.kamada_kawai_layout(G_thresholded)

    # Draw nodes
    node_sizes = [G_thresholded.degree(n) * 100 for n in G_thresholded]
    nx.draw_networkx_nodes(G_thresholded, pos, node_size=node_sizes, alpha=0.6, ax=axes[i])

    # Draw edges
    edges, weights = zip(*nx.get_edge_attributes(G_thresholded, 'weight').items())
    nx.draw_networkx_edges(G_thresholded, pos, edge_color='blue', alpha=0.5, width=1.0, ax=axes[i])

    # Draw labels
    nx.draw_networkx_labels(G_thresholded, pos, font_size=8, ax=axes[i])

    # Set the title for each subplot
    axes[i].set_title(f'Threshold: {threshold}', fontsize=12)
    axes[i].axis('off')

# Adjust layout to fit everything
plt.tight_layout()
# Display the plot
```

### Network Maps, Oregon Wildfires and Counties

### Threshold: 0.5
- Dense graph, which means that when the threshold for including an edge is a correlation of 0.5 or higher
- Almost every node is connected to nearly every other node.
- Relatively high baseline correlation between most fires in different Oregon counties.
In practical terms, this level of threshold might not be very informative because it does not help in distinguishing between the strengths of relationships—it suggests a broad similarity across the entire network without highlighting specifics.

### Threshold: 0.85
- Graph becomes less dense
- The edges that remain represent stronger correlations between the counties.
- All counties are correlated at some level, but a few pairs of counties are particularly strong.
- Some counties appear more central, indicating they may be counties that are connected to others
- An educated guess could suggest it's one fire spreading over multiple county lines, or discrepancies in data

### Threshold: 0.99
- Sparsest graph, indicating that very few counties have extremely high correlation. (note, in original network mapping, at a high threshold three clusters emerged)
- Clusters of counties indicate that they are strongly correlated with each other with respect to different variable measures for wildfires than with the rest of the network. 
  - May mean types of counties have a or some specific common feature or are influenced by similar factors.
- Additional information about the features of these specific counties may merit more investigation to understand why they correlate so strongly with others.

### General Observations:
- Edge Density: As the threshold increases, the number of edges decreases, focuses on the most significant relationships.
- Cluster Formation: At high thresholds, the formation of clusters can point to counties that share similarities of wildfire variables outlined in this dataset
